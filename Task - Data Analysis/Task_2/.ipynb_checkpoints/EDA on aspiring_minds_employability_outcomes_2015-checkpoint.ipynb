## Step - 1

The dataset was released by Aspiring Minds from the Aspiring Mind Employment Outcome 2015 (AMEO). The study is primarily limited  only to students with engineering disciplines. The dataset contains the employment outcomes of engineering graduates as dependent variables (Salary, Job Titles, and Job Locations) along with the standardized scores from three different areas – cognitive skills, technical skills and personality skills. The dataset also contains demographic features. The dataset  contains  around  40 independent variables and 4000 data points. The independent variables are both continuous and categorical in nature. The dataset contains a unique identifier for each candidate. Below mentioned table contains the details for the original dataset.

## Step - 3 Univariate Analysis

```python
sns.countplot(data['Gender'])
plt.show()
```

There are more male placed as compare to female

```python
desig = data['Designation'].value_counts()
arr_desig = list(desig.index)
arr_count = np.array(desig.values)
df = pd.DataFrame(data = arr_count.reshape(-1,1),columns=['Count'])
df['Designation'] = arr_desig

plt.bar(x = df['Designation'].iloc[:15] , height = df['Count'].iloc[:15])
plt.title('Most Designation')
plt.xticks(rotation = 90)
plt.show()
```

Most placed has role of software engineer

```python
ten = data['10percentage']
twelve = data['12percentage']
fig = plt.figure()
ax = fig.add_subplot(111)
ax.boxplot([ten,twelve], labels=['10percentage', '12percentage'])
```

we can clearly compare mean between marks of 10th and 12th standard

## Research Question 1:

Times of India article dated Jan 18, 2019 states that “After doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh
graduate.”

```python
x1 = data[(data['Designation'] == 'programmer analyst') & (data['GraduationYear'] == 2014)].Salary.values
x2 = data[(data['Designation'] == 'software engineer') & (data['GraduationYear'] == 2014)].Salary.values
x3 = data[(data['Designation'] == 'hardware engineer') & (data['GraduationYear'] == 2014)].Salary.values
x4 = data[(data['Designation'] == 'associate engineer') & (data['GraduationYear'] == 2014)].Salary.values
```

```python
sample = np.concatenate((x1,x2,x3,x4),axis=0)
sample_mean = np.mean(sample)

sample_mean
```

sample mean is greater than 3 lakhs so we can conclude that above statement is true

## Research Question 2:

Is there a relationship between gender and specialisation?

(i.e. Does the preference of Specialisation depend on the Gender?).”

```python
data['Gender'].value_counts()
```

```python
data['Specialization'].value_counts()
```

```python
observed = pd.crosstab(data.Specialization, data.Gender, margins=True)
observed.columns = ['Female','Male','All']
observed
```

from this table we conclude that there is no relationship between gender and specialisation
